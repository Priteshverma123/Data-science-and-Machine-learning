```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
 df = df.where((pd.notnull(df)),  '  ')
```

```python
X = df['Message']
Y= df['Category']
```

```python
le = preprocessing.LabelEncoder()
```

```python
# Y['Category']= le.fit_transform(Y['Category'])
```

```python
Y = le.fit_transform(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1 , stop_words = 'english' , lowercase = 'True')
```

```python
X_train= X_train.astype('str')
X_test= X_test.astype('str')
```

```python
X_train.info()
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features , Y_train)
```

```python
prediction1= model.predict(X_train_features)
accuracy = accuracy_score(Y_train , prediction1)
```

```python
print("accuracy on training data : " , accuracy )
```

```python
prediction2 = model.predict(X_test_features)
print(accuracy_score(Y_test , prediction2))
```

```python
# Calculation of Mean Squared Error (MSE)  
from sklearn.metrics import mean_squared_error
mse =mean_squared_error(Y_test,prediction2)  
print(mse)
```

```python
import math
rmse = math.sqrt(mse)
print("Root mean squared mean error of logistic regression: " , rmse)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(random_state = 0)
```

```python
dt.fit(X_train_features , Y_train)
```

```python
dtprediction1 = dt.predict(X_train_features)
print(accuracy_score(Y_train , dtprediction1))
```

```python
dtprediction2 = dt.predict(X_test_features)
print(accuracy_score(Y_test, dtprediction2))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
  # Calculation of Mean Squared Error (MSE)  
mse2 = mean_squared_error(Y_test,dtprediction2) 
print(mse2)
```

```python
rmse2 = math.sqrt(mse2)
print("Root mean squared mean error of decision tree :",rmse2)
```

```python
from xgboost import XGBClassifier
xgb= XGBClassifier()
```

```python
xgb.fit(X_train_features , Y_train)
```

```python
#for train data
xgbpred1 = xgb.predict(X_train_features)
print(accuracy_score(Y_train,xgbpred1))
```

```python
#for test data
xgbpred2 = xgb.predict(X_test_features)
print(accuracy_score(Y_test,xgbpred2))
```

```python
rmse2 = math.sqrt(xgbmse)
print("Root mean squared mean error of decision tree :",rmse2)
```
